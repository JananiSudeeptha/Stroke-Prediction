```python
import pandas as pd
import math
from apyori import apriori
```

```python
data = pd.read_csv("stroke_data.csv")
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data.head()
```

# Frequent Data set minig for stroke

```python
strokeData = data[data["stroke"] == 1]
strokeData = strokeData.replace({'sex': {1.0:'Male', 0.0:'Female'}})
strokeData = strokeData.replace({'ever_married': {0:'Not_married', 1:'Married'}})
strokeData = strokeData.replace({'hypertension': {0:'No_hypertension', 1:'Hypertension'}})
strokeData = strokeData.replace({'heart_disease': {0:'No_heart_disease', 1:'Heart_disease'}})
strokeData = strokeData.replace({'work_type': {0:'Never_worked', 1:'children', 2:'Govt_job', 3:'Self-employed', 4:'Private'}})
strokeData = strokeData.replace({'Residence_type': {0:'Rural', 1:'Urban'}})
strokeData = strokeData.replace({'smoking_status': {0:'never_smoked', 1:'smokes'}})
strokeData["age_category"] = strokeData["age"].map(lambda x : str(math.floor(x/10)*10) + "_s")
strokeData.head()
```

```python
columns = [column for column in strokeData.select_dtypes(include=['object'])]
strokeData = strokeData.loc[:,columns]
strokeData.head()
```

```python
association_rules = apriori(strokeData.values.tolist(), min_support=0.5, min_confidence=0.5)
association_results = list(association_rules)
for i in range(0,len(association_results)) :
    values = list(association_results[i][0])
    if len(values)>1:
        print(values)
```
