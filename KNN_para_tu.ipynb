```python
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score

# Define the hyperparameters to tune
space = {
    'n_neighbors': hp.choice('n_neighbors', range(1, 20)),
    'weights': hp.choice('weights', ['uniform', 'distance']),
    'metric': hp.choice('metric', ['euclidean', 'manhattan', 'chebyshev', 'minkowski'])
}

# Define the objective function to minimize
def objective(params):
    knn_model = KNeighborsRegressor(**params)
    score = cross_val_score(knn_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse = np.sqrt(-score.mean())
    return {'loss': rmse, 'status': STATUS_OK}

# Set up the trials object for tracking progress
trials = Trials()

# Run the hyperparameter search
best = fmin(fn=objective, space=space, algo=tpe.suggest, max_evals=50, trials=trials)

# Print the best hyperparameters found
print(best)

# Train the KNN model with the best hyperparameters on the full training set
best_params = space_eval(space, best)
knn_model = KNeighborsRegressor(**best_params)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("Test RMSE with best hyperparameters:", rmse)
```
